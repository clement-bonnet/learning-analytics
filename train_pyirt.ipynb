```python
def proba(question_id, user_id, item_param=item_param, user_param=user_param):
    item = item_param[question_id]
    alpha, beta, c = item["alpha"], item["beta"], item["c"]
    theta = user_param[user_id]
    return c + (1-c)/(1+np.exp(-alpha*(theta + beta)))
def apply_proba(r):
    return proba(r["QuestionId"], r["UserId"])
```

```python
df_pred = df.copy()
df_pred["Pred"] = df_pred.apply(apply_proba, axis=1)
```

```python
THRESHOLD = 0.5
df_pred["Pred_bin"] = df_pred.apply(lambda r: int(r["Pred"]>THRESHOLD), axis=1)
```
